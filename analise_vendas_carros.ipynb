### Construindo um modelo de machine learning

```python
df.head()
```

##### Pré-processamento dos dados

```python
df.drop('vehicle.age.group', axis = 1, inplace = True)#Apago essa coluna porque ela é redundante. axis 1 porque é coluna e não linha
```

```python
df.drop('data.set', axis = 1, inplace = True)#apago porque nao agrega ao modelo
```

##### Aplicando o label encoder

```python
from sklearn.preprocessing import LabelEncoder#importando
```

```python
#criei 6 objetos do tipo label_encoder
label_encoder1 = LabelEncoder()
label_encoder2 = LabelEncoder()
label_encoder3 = LabelEncoder()
label_encoder4 = LabelEncoder()
label_encoder5 = LabelEncoder()
label_encoder6 = LabelEncoder()
```

```python
#Cada coluna vai receber seu label encoder que vai codificar categorica para numerica
df['vehicle.type'] = label_encoder1.fit_transform(df['vehicle.type'])
df['domestic.import']= label_encoder2.fit_transform(df['domestic.import'])
df['color.set']= label_encoder3.fit_transform(df['color.set'])
df['makex']= label_encoder4.fit_transform(df['makex'])
df['state']= label_encoder5.fit_transform(df['state'])
df['make.model']= label_encoder6.fit_transform(df['make.model'])
```

```python
df.head()
```

#### Separa a classe dos dados

```python
y = df['overage']
y.head()
```

#### Apaga a coluna overage e a coluna lot.days.sales

```python
X = df.drop(['lot.sale.days', 'overage'], axis = 1)
```

```python
X.head()
```

#### Separando dados de treino e de teste

```python
from sklearn.model_selection import train_test_split#importando. Ele separa dados de treino e teste de forma aleatoria
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y)#Crio as 4 variaveis com X treino e teste e y treino e teste. 75% para treino e 25% para teste
```

```python
X_treino.count()
```

```python
y_treino.count()
```

```python
X_teste.count()
```

```python
y_teste.count()
```

### Aplicando Machine Learning com árvore de decisão

```python
from sklearn import tree #vou fazer uma classificação para saber se um novo carro vai sofrer overage ou não
```

```python
arvore = tree.DecisionTreeClassifier()#instanciei o algoritmo
```

```python
arvore.fit(X_treino, y_treino)
```

### Validação do modelo

```python
resultado = arvore.predict(X_teste)
```

```python
resultado#vendo o que meu modelo previu e que está salvo na variavel resultado
```

```python
from sklearn import metrics #vai contabilizar minha quantidade de acertos
print(metrics.classification_report(y_teste, resultado))#passo dois parametros. O Y teste, que são dados reais, e o resultado que foi previsto
```

### Features mais importantes para o modelo

```python
arvore.feature_importances_ #Esse atributo do meu modelo dá os pesos para cada feature
```

```python
feature_imp = pd.Series(arvore.feature_importances_, index = X_treino.columns)#Crio uma series salvando os pesos. O index é o nome da coluna do X_treino
```

```python
feature_imp
```

### Visualizando as features mais importantes

```python
%matplotlib notebook
sns.barplot(x = feature_imp, y = feature_imp.index)#passo a importancia como x e nome como Y
plt.xlabel('Importância das features')
plt.ylabel('Features')
plt.title('Importância das Features')
plt.show()
```
